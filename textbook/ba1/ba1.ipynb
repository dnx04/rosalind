## [B](https://rosalind.info/problems/ba1b/)

```python
fi = open('input/rosalind_ba1b.txt', 'r')

d = {}
s, k = fi.read().split()
k = int(k)

for i in range(len(s) - k + 1):
  d[s[i:i+k]] = d.get(s[i:i+k], 0) + 1

pick = max(d, key=d.get)
for key, val in d.items():
  if val == d[pick]:
    print(key)
```

## [D](https://rosalind.info/problems/ba1d/)

## [E](https://rosalind.info/problems/ba1e/)

```python
fi = open('input/rosalind_ba1e.txt', 'r')

genome = fi.readline()
k, l, t = map(int, fi.readline().split())

d = {}
for i in range(len(genome) - k + 1):
    d[genome[i:i+k]] = d.get(genome[i:i+k], 0) + 1

clumps = []
for key, val in d.items():
    if val >= t:
        clumps.append(key)

for c in clumps:
    print(c, end=' ')
```

## [F](https://rosalind.info/problems/ba1f/)

```python
fi = open('input/rosalind_ba1f.txt', 'r')

genome = fi.readline()
genome = Seq(genome)

res = [0]
mn, cnt = 0, 0
for i in range(len(genome) - 1):
    if genome[i] == 'G':
        cnt += 1
    elif genome[i] == 'C':
        cnt -= 1
    if mn >= cnt:
        if mn > cnt:
            mn = cnt
            res = [i]
        else:
            res.append(i)

print(' '.join([str(i + 1) for i in res]))
```
